```python
import requests
import xmltodict
import json

# 공공데이터포털 사이트로 크롤링 요청 위한 인증키와 요청 메세지
key = "y3hJyZd4gVr4xFzmKzqb53nioC68jdCSKnf8lZYuhvp9VOgSJ5N1GgtPP7fEVhcasTygTCSvigiqKj2syoMEyg%3D%3D"
url_cycle="http://openapi.jejusi.go.kr/rest/PublicBikeInfoService/getPublicBikeInfoList?ServiceKey={}&pageNo=1&numOfRows=10".format(key)
url_parking = "http://openapi.jejuits.go.kr/OPEN_API/pisInfo/getPisInfo?serviceKey={}&pageNo=1&numOfRows=100".format(key)

# 웹에 요청
content_cycle = requests.get(url_cycle).content
content_parking = requests.get(url_parking).content
# xml형식은 정보 분리하기 힘들어서 json으로 파싱
dict_cycle = xmltodict.parse(content_cycle)
dict_parking = xmltodict.parse(content_parking)
# json으로 변화된 정보를 구분해서 출력-공공자전거
jsonString_cycle = json.dumps(dict_cycle['rfcOpenApi']['body'], ensure_ascii = False)
jsonObj_cycle = json.loads(jsonString_cycle)

for list in jsonObj_cycle['data']['list']:
    print(list['address'],list['location'],list['retal_enable_num'])

print('\n')

# json으로 변화된 정보를 구분해서 출력-주차장
jsonString_parking = json.dumps(dict_parking['ServiceResult']['msgBody'], ensure_ascii = False)
jsonObj_parking = json.loads(jsonString_parking)

for list in jsonObj_parking['itemList']:
    print(list['ISTL_LCTN_ADDR'],list['WHOL_NPLS'],list['GNRL_RMND_PRZN_NUM'],list['HNDC_RMND_PRZN_NUM'])

print("------------------------------------자전거 관련 코드 -----------------------")
    
import pandas as pd
import matplotlib as plt
import folium

idx_col = 'category' # 인덱스를 종류구분체계 열로 지정
df =pd.read_csv('jeju_info.csv',header=0,encoding='CP949',index_col=idx_col)


print("---------------------------------------------------------------")
pd.set_option('display.width',None)
pd.set_option('display.max_rows',100)
pd.set_option('display.max_columns',10)
pd.set_option('display.max_colwidth',20)
pd.set_option('display.unicode.east_asian_width',True)

print(df.columns.values)
# 지도에 위치 표시
jeju_map = folium.Map(location=[33.389398,126.541236], tiles='stamen Terrain',
                     zoom_start=11)





#-----------------------------------------변경할 부분 ----------------------------------------- 





#출력한 지명 입력
cluster = input("지도에 출력할 분류명 입력 : ")


# 분류명 종류 -> 공공기관 , 공원/산/동.식물원 , 관광/숙박 , 교육기관 , 교통시설 , 금융기관 , 
#  기업 , 농공시설 , 도로시설 , 레져/스포츠 , 문화/종교/예술 , 쇼핑/편의 , 음식점 , 의료/복지 
#  자동차관련 , 지명관련 , 치안관련 

# input 부분을 지우고 "분류명"들을  -> 콤보박스로 만들어서
# 콤보박스 클릭시 그 "분류명"을 => cluster라는 이름의 변수에 저장하게 하는 코드.





#-----------------------------------------변경할 부분 -----------------------------------------

#데이터프레임에 특정 컬럼만 저장
df_sample = df.loc[:,['category','loc_name','loc_x','loc_y']]  
text = "df_sample.loc[['"+cluster+"'],:]"
#특정 행만 검색
df_sample2 = df_sample.loc[[cluster],:]

# 결과 리스트 출력
print('\n전체 데이터 수')
print(df_sample)
print('\n 해당 데이터 수')
print(df_sample2)

# 지도에 마커 출력
for name, lat, lng in zip(df_sample2.loc_name, df_sample2.loc_x, df_sample2.loc_y):
    folium.Marker([lat,lng],
                         radius=2,
                         color='brown',
                         fill=True,
                         fill_color='coral',
                         fill_opacity=0.7,
                         popup=name
                       ).add_to(jeju_map)

# HTML 파일로 저장    
jeju_map.save('./jeju_locationEX.html')
```
